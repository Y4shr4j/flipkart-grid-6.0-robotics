```python
  # Install Tesseract and the required Python libraries
  !sudo apt-get install -y tesseract-ocr
  !pip install pytesseract
  !pip install opencv-python-headless

  # Import necessary libraries
  import cv2
  import pytesseract
  from PIL import Image
  import numpy as np
  from google.colab import files

  # Function to preprocess the image for better OCR results
  def preprocess_image(image_path):
      # Load image using OpenCV
      image = cv2.imread(image_path)

      # Convert the image to grayscale
      gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

      # Apply thresholding to clean up the image
      _, binary_image = cv2.threshold(gray_image, 150, 255, cv2.THRESH_BINARY_INV)

      # Optionally, apply some morphological operations to remove noise
      kernel = np.ones((1, 1), np.uint8)
      cleaned_image = cv2.morphologyEx(binary_image, cv2.MORPH_CLOSE, kernel)

      return cleaned_image

  # Function to extract text using OCR
  def perform_ocr(image_path):
      # Preprocess the image
      preprocessed_image = preprocess_image(image_path)

      # Convert the preprocessed image back to PIL format for pytesseract
      pil_image = Image.fromarray(preprocessed_image)

      # Use pytesseract to extract text from the image
      extracted_text = pytesseract.image_to_string(pil_image)

      return extracted_text

  # Upload the image file
  uploaded = files.upload()
  image_path = list(uploaded.keys())[0]  # Get the uploaded image file path

  # Perform OCR on the uploaded image
  extracted_text = perform_ocr(image_path)

  # Output the extracted text
  print("Extracted Text:")
  print(extracted_text)
```
